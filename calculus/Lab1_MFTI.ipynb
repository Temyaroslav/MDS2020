```python
def makeData():
    x = np.arange(-10, 10, 0.1)
    y = np.arange(-10, 10, 0.1)
    xgrid, ygrid = np.meshgrid(x, y)
    zgrid = np.sin(xgrid)*np.sin(ygrid)/(xgrid*ygrid)
    return xgrid, ygrid, zgrid

x, y, z = makeData()

fig = plt.figure()
axes = axes3d.Axes3D(fig)
axes.plot_surface(x, y, z, rstride=4, cstride=4, cmap=cm.jet)
plt.show()
```

### Ex 3

```python
def func(x):
    y = np.log((x**2 + 1) * (np.exp(-abs(x) / 10))) / np.log(1 + np.tan(1 / (1 + np.sin(x)**2)))
    return y

x = np.arange(-10, 10, 0.1)

plt.plot(x, func(x))
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.show()
```

### Ex 5

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 1.42, 1.76, 2, 2.24, 2.5]

p, v = np.polyfit(x, y, deg=1, cov=True)
p_f1 = np.poly1d(p)
p, v = np.polyfit(x, y, deg=2, cov=True)
p_f2 = np.poly1d(p)

plt.plot(x, p_f1(x), c='C2', label=r'$P_1(x)$')
plt.plot(x, p_f2(x), c='C3', label=r'$P_2(x)$')
plt.errorbar(x, y, c='C0', xerr=0.05, yerr=0.1)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$P_i(x)$', fontsize=14)
plt.legend(loc='best', fontsize=12)
plt.grid()
plt.show()
```

### Ex 6

```python
b = 11
a = (1 + 3/2 * np.pi) / b

N = 300
x = np.arange(-2, 2, 0.01)

def func(x, N):
    a_n = np.power(np.full((1, N), a), np.arange(0, N, 1))
    b_n = np.power(np.full((1, N), b), np.arange(0, N, 1))
    s = a_n @ np.cos(b_n * np.pi * x).T
    return s.ravel()[0]

y = [func(i, N) for i in x]

plt.figure(figsize=(10, 8))
plt.plot(x, y)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.title('Weierstrass function')
plt.show()
```
