```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
sample_size = 10
pop_mean = 0.5 # for uniform distrib in [0, 1]
iterations = int(10e3)

t_values = []
for i in range(iterations):
    sample = np.random.uniform(0, 1, sample_size)
    t = (np.mean(sample) - pop_mean) / np.sqrt(np.var(sample, ddof=1)) * np.sqrt(sample_size)
    t_values.append(t)
```

```python
x = np.linspace(-7, 7)

plt.hist(t_values, bins=30, density=True, label='t-distrib')
plt.plot(x, scipy.stats.norm.pdf(x), label='norm')
plt.legend()
plt.xlim(-7, 7)
plt.show()
```

t-statistic doesn't really follow Normal distribution, which can be seen in the tails. t-statistics follows T distribution instead.

```python
x = np.linspace(-7, 7)

plt.hist(t_values, bins=30, density=True, label='t-distrib')
plt.plot(x, scipy.stats.t.pdf(x, df=sample_size - 1), label='t-dstrib (theo)')
plt.legend()
plt.xlim(-7, 7)
plt.show()
```

Gives better approximation to the hist we got from the simulation

```python
x = np.linspace(-7, 7)

for i in [1, 5, 20]:
    plt.plot(x, scipy.stats.t.pdf(x, df=i), label=f't-dstrib (df={i})')
plt.plot(x, scipy.stats.norm.pdf(x), label='norm')


plt.legend()
plt.xlim(-7, 7)
plt.show()
```
